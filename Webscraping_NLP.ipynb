```python
#Crawl function
import requests
from bs4 import BeautifulSoup

def web_page_crawler(url, attributes):

  final_data = ""
  try:
      result = requests.get(URL)
      soup = BeautifulSoup(result.content, 'html.parser')
      soup_content = soup.find("div", attrs = attributes)
      final_data = soup_content.text
  except Exception as e:
      print("error", e)

  return final_data
```

```python
class LogisticClassifier():
    def __init__(self):
        self.labels = list(filter(lambda x: not (x.endswith("~")), os.listdir("./train_data")))
        self.vectorizer = TfidfVectorizer( max_features = None, strip_accents = 'unicode',
                            analyzer = "word", ngram_range=(1,2), use_idf = 1, smooth_idf = 1, tokenizer=self.tokenize, stop_words='english') 
    
    def tokenize(self, text):
        text = text.lower()
        tokens = []
        sent_detector = nltk.data.load('tokenizers/punkt/english.pickle')
        stop_words = {}
        stop_words_punct = [s for s in string.punctuation]
        text = re.sub(r'\.',' ', text)
        text = re.sub(r'\s+',' ', text)
        sents = sent_detector.tokenize(text.strip())
        # print sents[0:1]
        for stext in sents[0:1]:
            tokens += wordpunct_tokenize(stext)
        return tokens
    
    def getData(self):
        xdata = []
        ydata = []
        train_path = './train_data'
        try:
            for ifile in self.labels:
                with open(os.path.join(train_path,ifile),"r") as tf:


                    tmpdata = tf.readlines()
                tmpdata = [ i.replace("\n","") for i in tmpdata ]
                tmpdata = [ i for i in tmpdata if i != "" ]
                for i in tmpdata:
                    xdata.append(i)
                    ydata.append(self.labels.index(ifile))
        except:
            import traceback
            print (traceback.format_exc())
        
        return xdata, ydata
        
    """
    Function to train classifier
    """
    def train(self):
        X_train, Y_train = self.getData()
        vectors_train = self.vectorizer.fit_transform(X_train)
        self.classifier = OneVsRestClassifier( LogisticRegression( C = 2.0, multi_class = "ovr", solver = "liblinear" ) )
        self.classifier.fit(vectors_train, Y_train)

    """
    Function to test classifier
    """
    def test(self, query):
        modifier_dict = {"collaboration.txt":"collaboration",
                         "history.txt":"history",
                         "vision.txt":"vision"
           
        }
        self.train()
        classifier_result = {}
        vectors_test = self.vectorizer.transform([query])
        pred = self.classifier.predict(vectors_test)
        predp = self.classifier.predict_proba(vectors_test)
        predp = predp.tolist()[0]
        max_ind =  predp.index(max(predp))
        classifier_result["class"] = modifier_dict[self.labels[max_ind]]
        classifier_result["prediction"] = {self.labels[max_ind]: round(max(predp)*100.0, 2)*0.6}
        
        return classifier_result


            
            
result = LogisticClassifier()
d = result.test(str(sentence))
print (d)  


```

```python
text = all_data["vision"] + all_data["history"] + all_data["collaborations"]
text_list = text.split(".")

```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer(max_df=0.8, min_df=1, stop_words='english')
doc_term_matrix = count_vect.fit_transform(text_list)
```

```python
doc_term_matrix 
```

```python
from sklearn.decomposition import LatentDirichletAllocation #probabilistic

LDA = LatentDirichletAllocation(n_components=4, random_state=42)
LDA.fit(doc_term_matrix)
```

```python
first_topic = LDA.components_[0]
first_topic
top_topic_words = first_topic.argsort()[-10:]

for i in top_topic_words:
    print(count_vect.get_feature_names()[i])
```

```python
!python3 -m spacy download en_core_web_md

import spacy
nlp = spacy.load('en_core_web_md')
```

```python
# focus_words = ["abhiyan"]
import enchant
all_score = []

for i,topic in enumerate(LDA.components_):

    topics = [count_vect.get_feature_names()[i] for i in topic.argsort()[-10:]]

    _score = 0
    
    for each_word in focus_words:
      for each_topic_word in topics:
        doc1 = nlp(each_word)
        doc2 = nlp(each_topic_word)
        _score += doc1.similarity(doc2)
        print("matching score", each_word, each_topic_word, _score)

    print('Top 10 words for topic #', i, '\n', topics, _score, "\n\n")
    all_score.append(_score)
    _score = 0

    
```

```python
topic_values = LDA.transform(doc_term_matrix)
print(topic_values.shape)
print(topic_values.argmax(axis=1))
```

```python
print(all_score)

m = max(all_score)
all_indexes = [i for i, j in enumerate(all_score) if j == m]
print(all_indexes)
answer_list = []

```

```python
for each_index in all_indexes:
  for i, each_topic_number in enumerate (list(topic_values.argmax(axis=1))):
    if each_topic_number == each_index:
      answer_list.append(text_list[i])
      
". ".join(answer_list).strip()
```
